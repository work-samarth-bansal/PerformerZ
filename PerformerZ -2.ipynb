# Promotion in Last Five Years

```python
summary_stats = df['promotion_last_5years'].describe()
summary_stats
```

```python
print(df.corr()['promotion_last_5years'])
```

```python
promotions_last_5_years = df['promotion_last_5years'].sum()
total_employees = len(df)
percentage_promoted = (promotions_last_5_years / total_employees) * 100
print("Number of employees promoted in the last 5 years:", promotions_last_5_years)
print("Percentage of employees promoted in the last 5 years: {:.2f}%".format(percentage_promoted))
```

```python
promotions_last_5_years = df['promotion_last_5years'].sum()
total_employees = len(df)
percentage_promoted = (promotions_last_5_years / total_employees) * 100
plt.figure(figsize=(6, 6))
plt.bar([' Not Promoted', 'Promoted'], [promotions_last_5_years, percentage_promoted], color=['brown', 'orange'])
plt.title("Promotions in the Last 5 Years")
plt.ylabel("Value")
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
sns.histplot(df['promotion_last_5years'], kde=True)
plt.title('Promotions During Last 5 Years')
plt.xlabel('Promotion')
plt.ylabel('Frequency')
plt.show()
```

```python
promotions_by_department = df.groupby('sales')['promotion_last_5years'].value_counts(normalize=True).unstack()
print("\nPromotion Status by Department:")
print(promotions_by_department)
```

```python
promotions_by_department = df.groupby('sales')['promotion_last_5years'].value_counts(normalize=True).unstack()
promotions_by_department.plot(kind='bar', stacked=True, colormap='Paired')
plt.title("Promotion Status by Department")
plt.xlabel("Department")
plt.ylabel("Percentage")
plt.legend(title='Promotion Status', labels=['Not Promoted', 'Promoted'])
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
promotions_by_salary = df.groupby('salary')['promotion_last_5years'].value_counts(normalize=True).unstack()
print("\nPromotion Status by Salary Level:")
print(promotions_by_salary)
```

```python
promotions_by_salary = df.groupby('salary')['promotion_last_5years'].value_counts(normalize=True).unstack()
custom_order = ['high', 'medium', 'low']
promotions_by_salary = promotions_by_salary.reindex(custom_order)
promotions_by_salary.plot(kind='bar', stacked=True, colormap='Paired', figsize=(8, 6))
plt.title("Promotion Status by Salary Level")
plt.xlabel("Salary Level")
plt.ylabel("Percentage")
plt.legend(title='Promotion Status', labels=['Not Promoted', 'Promoted'])
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
promotions_satisfied = df[df['promotion_last_5years'] == 1]['satisfaction_level'].median()
promotions_not_satisfied = df[df['promotion_last_5years'] == 0]['satisfaction_level'].median()
print("\nAverage Satisfaction Level of Promoted Employees: {:.2f}".format(promotions_satisfied))
print("Average Satisfaction Level of Non-Promoted Employees: {:.2f}".format(promotions_not_satisfied))
```

```python
promotions_satisfied = df[df['promotion_last_5years'] == 1]['satisfaction_level'].median()
promotions_not_satisfied = df[df['promotion_last_5years'] == 0]['satisfaction_level'].median()
satisfaction_levels = [promotions_satisfied, promotions_not_satisfied]
labels = ['Promoted', 'Not Promoted']
plt.figure(figsize=(6, 6))
plt.bar(labels, satisfaction_levels, color=['red', 'green'])
plt.title("Average Satisfaction Level of Promoted vs. Non-Promoted Employees")
plt.ylabel("Average Satisfaction Level")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
time_in_company_promotion_rate = df.groupby('time_spend_company')['promotion_last_5years'].mean()
print("Promotion Rate by Time in the Company:")
print(time_in_company_promotion_rate)
```

```python
time_in_company_promotion_rate = df.groupby('time_spend_company')['promotion_last_5years'].mean()
plt.figure(figsize=(8, 6))
ax = time_in_company_promotion_rate.plot(kind='bar', color='skyblue')
plt.title("Promotion Rate by Time in the Company")
plt.xlabel("Time in the Company (Years)")
plt.ylabel("Promotion Rate")
plt.grid(axis='y', linestyle='--')
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
plt.show()
```

```python
performance_evaluation_promotion_rate = df.groupby('promotion_last_5years')['Work_accident'].mean()
print("Promotion Rate by Performance Evaluation:")
print(performance_evaluation_promotion_rate)
```

```python
performance_evaluation_promotion_rate = df.groupby('Work_accident')['promotion_last_5years'].mean()
plt.figure(figsize=(12, 6))
ss=performance_evaluation_promotion_rate.plot(kind='bar', color='skyblue')
plt.title("Promotion Rate by Performance Evaluation")
plt.xlabel("Performance Evaluation")
plt.ylabel("Promotion Rate")
plt.grid(axis='y', linestyle='--')
ss.set_xticklabels(ss.get_xticklabels(), rotation=0)
plt.show()
```
