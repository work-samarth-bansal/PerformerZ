# Predicting Left

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import VotingClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from imblearn.over_sampling import SMOTE

# Load the dataset
# Assuming you have your dataset loaded into a variable called 'dataf'

# Preprocessing
dtf = dataf.copy()

# Feature Engineering
# You can perform feature engineering here if needed

# Handling Categorical Variables (One-Hot Encoding for 'sales' column)
dtf = pd.get_dummies(dtf, columns=['sales', 'salary'], drop_first=True)

# Split data into features (X) and target variable (y)
X = dtf.drop(['left'], axis=1)
y = dtf['left']

# Data Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Balancing the Dataset using SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

# Train logistic regression model
model1 = LogisticRegression()
model2 = SVC()
model3 = DecisionTreeClassifier()

voting_clf = VotingClassifier(estimators=[('lr', model1), ('svc', model2), ('dt', model3)], voting='hard')
voting_clf.fit(X_resampled, y_resampled)

# Evaluate model performance
print("Accuracy on test set: {:.2f}".format(voting_clf.score(X_test, y_test)))

# Sample input data for prediction
input_data = [0.2, 0.45, 2, 140, 3, 0, 1]

# Standardize input data and make prediction
input_data_scaled = scaler.transform([input_data])
probability = voting_clf.predict_proba(input_data_scaled)[0][1]

print("Probability of leaving:", probability)
```

# Predicting Satisfaction Level

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

df3 = dataf.copy()
df3 = df3.drop(['sales'], axis=1)
df3 = df3.drop(['salary'], axis=1)

# Define the features and the target variable
X = df3.drop(['satisfaction_level'], axis=1)
y = df3['satisfaction_level']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create and train a Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Evaluate model performance using the R-squared metric
r_squared = model.score(X_test, y_test)
print("Accuracy on test set: {:.2f}".format(model.score(X_test, y_test)))

# Sample input data for prediction
input_data = [[0.2, 0.45, 2, 140, 3, 0, 1]]

# Predict satisfaction level
predicted_satisfaction = model.predict(input_data)
print("Probability of Satisfaction Level:", predicted_satisfaction[0])
```

```python
from sklearn.ensemble import RandomForestRegressor

# Create and train a Random Forest model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Evaluate model performance using the R-squared metric
r_squared = model.score(X_test, y_test)
print("Accuracy on test set: {:.2f}".format(model.score(X_test, y_test)))

# Sample input data for prediction
input_data = [[0.2, 0.45, 2, 140, 3, 0, 1]]

# Predict satisfaction level
predicted_satisfaction = model.predict(input_data)
print("Probability of Satisfaction Level:", predicted_satisfaction[0])
```
